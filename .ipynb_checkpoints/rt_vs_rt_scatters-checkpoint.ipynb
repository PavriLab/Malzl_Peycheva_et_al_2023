```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
import numpy as np
import matplotlib
import pyBigWig as pbw
from scipy.stats import binned_statistic_2d
from matplotlib.collections import LineCollection
from scipy.interpolate import interp1d

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

def bw_to_df(bwfile, chroms, valuename, binsize = 20000):
    bw = pbw.open(bwfile)
    df = pd.DataFrame(
        columns = ['chrom', 'start', 'end', valuename]
    )
    for chrom in chroms:
        intervals = []
        for start, end, value in bw.intervals(chrom):
            if (end - start) > binsize:
                intervals.extend(
                    [(chrom, pos, pos + binsize, value) for pos in range(start, end, binsize)]
                )
            
            else:
                intervals.append((chrom, start, end, value))
            
        chrom_df = pd.DataFrame(
            intervals,
            columns = ['chrom', 'start', 'end', valuename]
        )
        df = pd.concat(
            [df, chrom_df]
        )
    bw.close()
    return df


rtfiles = {
    'shLacZ_tc': '../../RepliSeq/mcmkd_timecourse/repliseq-nf/results20kb/bedGraph/timecourse.WT.loess.bedGraph',
    'shMcm_d5': '../../RepliSeq/mcmkd_timecourse/repliseq-nf/results20kb/bedGraph/timecourse.Mcm6kd.d5.loess.bedGraph',
    'Rif1ko_tc': '../../RepliSeq/mcmkd_timecourse/repliseq-nf/results20kb/bedGraph/timecourse.Rif1ko.loess.bedGraph',
    'Rif1ko_shMcm_d5': '../../RepliSeq/mcmkd_timecourse/repliseq-nf/results20kb/bedGraph/timecourse.Mcm6kdRif1ko.d5.loess.bedGraph',
    'WT': '/groups/pavri/bioinfo/daniel/buonomo/RT/results20kb/bedGraph/WT.bedGraph',
    'Rif1Het': '/groups/pavri/bioinfo/daniel/buonomo/RT/results20kb/bedGraph/Rif1KOsingle.bedGraph',
    'Rif1KO': '/groups/pavri/bioinfo/daniel/buonomo/RT/results20kb/bedGraph/Rif1KOdouble.bedGraph'
}

condition_pairs = [
    ('shLacZ_tc', 'shMcm_d5'),
    ('shLacZ_tc', 'Rif1ko_tc'),
    ('Rif1ko_tc', 'Rif1ko_shMcm_d5'),
    ('shMcm_d5', 'Rif1ko_shMcm_d5'),
    ('WT', 'Rif1Het'),
    ('WT', 'Rif1KO')
]

cmap = colors.LinearSegmentedColormap.from_list(
    'bla', 
    plt.get_cmap('coolwarm')(np.linspace(0.5, 1, 256))
)

n_bins = 150
xnorm = interp1d((-7, 7), (0, n_bins))
ynorm = interp1d((-7, 7), (0, n_bins))

cmap = colors.LinearSegmentedColormap.from_list(
    'bla', 
    plt.get_cmap('coolwarm')(np.linspace(0.5, 1, 256))
)
vmin, vmax = 0, 15
abs_ylims = [7] * len(condition_pairs)
abs_xlims = [7] * len(condition_pairs)
fig, axs = plt.subplots(1, len(condition_pairs))
for ax, (c1, c2), ylim, xlim in zip(axs, condition_pairs, abs_ylims, abs_xlims):
    d1 = pd.read_csv(
        rtfiles[c1],
        sep = '\t', 
        header = None, 
        names = ['chrom', 'start', 'end', c1]
    )
    d2 = pd.read_csv(
        rtfiles[c2],
        sep = '\t', 
        header = None, 
        names = ['chrom', 'start', 'end', c2]
    )
    df = d1.merge(
        d2,
        on = ['chrom', 'start', 'end'],
        how = 'left',
    )
    sns.histplot(
        data = df,
        x = c1,
        y = c2,
        binrange = ((-xlim, xlim), (-ylim, ylim)),
        bins = (150, 150),
        stat = 'count',
        cmap = cmap,
        ax = ax,
        cbar = True,
        vmax = 100
    )
    ax1.set_xlim(-xlim, xlim)
    ax1.set_ylim(-ylim, ylim)

    if c1 == 'WT':
        continue


fig.set_figwidth(6 * len(condition_pairs))
fig.set_figheight(5)
fig.tight_layout()
```
