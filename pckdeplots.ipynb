```python
import seaborn as sns
import pandas as pd

conditions = ['shLacZ', 'shMcm', 'Rif1ko', 'Rif1ko_shMcm']
eigs = {}
for c in conditions:
    eigs[c] = pd.read_csv(
        f'ditag/CH12_HiC_{c}/CH12_HiC_{c}_20kb_w200000.PC1.bedGraph',
        sep = '\t',
        skiprows = 1,
        header = None,
        names = ['chrom', 'start', 'end', c]
    )
eigs
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as colors
import numpy as np
mpl.rcParams['pdf.fonttype'] = 42

comparisons = [
    ('shLacZ', 'shMcm'),
    ('shLacZ', 'Rif1ko'),
    ('Rif1ko', 'Rif1ko_shMcm')
]

cmap = colors.LinearSegmentedColormap.from_list(
    'bla', 
    plt.get_cmap('coolwarm')(np.linspace(0.5, 1, 256))
)

lims = (-2, 2)

fig, axs = plt.subplots(2, len(comparisons))
for i, fill in enumerate([True, False]):
    for (c1, c2), ax in zip(comparisons, axs[i, :]):
        df = eigs[c1].merge(
            eigs[c2],
            on = ['chrom', 'start', 'end'],
            how = 'left'
        )
        
        if fill:
            sns.kdeplot(
                x = df[c1],
                y = df[c2],
                clip = (lims, lims),
                gridsize = 150,
                ax = ax,
                fill = fill,
                cmap = cmap
            )
        
        else:
            sns.kdeplot(
                x = df[c1],
                y = df[c2],
                clip = (lims, lims),
                gridsize = 150,
                ax = ax,
            )

        ax.set_xticks(np.linspace(*lims, 5))
        ax.set_yticks(np.linspace(*lims, 5))
        ax.axvline(0, c='k', lw=0.5, ls='--')
        ax.axhline(0, c='k', lw=0.5, ls='--')
        ax.plot(lims, lims , c='k', lw=0.5, ls='--')
        ax.set_xlim(*lims)
        ax.set_ylim(*lims)

        ax.set_ylabel(f'PC1 {c2}')
        ax.set_title(f'{c2} vs {c1}')
        ax.set_xlabel(f'PC1 {c1}')
    
fig.set_figwidth(3.75 * len(comparisons))
fig.set_figheight(8)
fig.tight_layout()
fig.savefig('pc1scatter.kde.pdf')
```
